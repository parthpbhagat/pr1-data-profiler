* perform univeriant 
   * Distribution plot

```python
df_csv["type"].value_counts()
```

```python
import matplotlib.pyplot as plt

movie_type = df_csv["type"].value_counts()

bar = plt.barh(movie_type.index, movie_type.values)
plt.bar_label(bar)
plt.xlabel("Type")
plt.ylabel("Category")
plt.title("Count of Each Type")
plt.show()
```

```python
import matplotlib.pyplot as plt

categarical = df_csv["release_year"].value_counts()

plt.bar(categarical.index,categarical.values)
plt.xlabel("Relese year")
plt.ylabel("Count")
plt.title("Count of movie per Release year ")
plt.show()
```

* Perform multivariant

```python
import matplotlib.pyplot as plt

rating = df_csv["rating"].value_counts()

bar = plt.bar(rating.index, rating.values)
plt.bar_label(bar)
plt.xlabel("Rating")
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.ylabel("Category")
plt.title("Rating of Each Categery")
plt.show()
```
